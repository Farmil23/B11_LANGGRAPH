```python
from langchain_groq import ChatGroq
import os
from dotenv import load_dotenv

from pydantic import BaseModel
from typing import List

load_dotenv()

api_key = os.getenv("GROQ_API_KEY")
model = ChatGroq(model_name="openai/gpt-oss-120b", api_key=api_key)
```

```python
model.invoke("test")
```

```python
class book(BaseModel):
    pembuat: str = Field(description="Nama pembuat buku")
    halaman: int = Field(description="Jumlah halaman buku")
    harga: int = Field(description="Harga buku")
    nama_buku: str = Field(description="Nama buku")

model_with_structured = model.with_structured_output(book)

response = model_with_structured.invoke("carikan aku buku tentang ai engineer dan tentang rag?")

```

```python
from pydantic import BaseModel, Field
from enum import Enum
from typing import List

# --- Definisi Model Pydantic (Dari Latihan 3) ---
class Sentiment(str, Enum):
    POSITIVE = "positive"
    NEGATIVE = "negative"
    NEUTRAL = "neutral"

class Category(str, Enum):
    PRODUCT_QUALITY = "product_quality"
    SHIPPING = "shipping"
    CUSTOMER_SERVICE = "customer_service"

class ReviewAnalysis(BaseModel):
    summary: str = Field(description="Ringkasan singkat ulasan dalam 1 kalimat")
    sentiment: Sentiment = Field(description="Sentimen keseluruhan ulasan")
    tags: List[Category] = Field(description="Daftar kategori yang relevan")
```

```python
# 3. Eksekusi
input_text = "Pelayanan CS sangat ramah membantu saya retur barang, walau prosesnya agak lama."

structured_llm = model.with_structured_output(ReviewAnalysis)
try:
    hasil = structured_llm.invoke(input_text)
    
    
    print(f"Type object: {type(hasil)}") 
    print("-" * 20)
    print(f"Ringkasan: {hasil.summary}")
    print(f"Kategori : {hasil.tags}")
    
except Exception as e:
    print(f"Error parsing: {e}")
```
